```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

def power_function(x, a, b):
    return a * x**b

n = int(input("Enter the number of data points: "))
x = [float(input(f"Enter x[{i+1}]: ")) for i in range(n)]
y = [float(input(f"Enter y[{i+1}]: ")) for i in range(n)]

params, _ = curve_fit(power_function, x, y)
a, b = params

plt.scatter(x, y, color='red', label='Data Points')
plt.plot(sorted(x), power_function(np.array(sorted(x)), a, b), color='blue', label=f'Best Fit: y = {a:.4f} * x^{b:.4f}')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Power Function Curve Fitting")
plt.legend()
plt.show()

print(f"The best-fit equation is: y = {a:.4f} * x^{b:.4f}")
```
