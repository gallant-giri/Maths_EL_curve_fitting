```python
import numpy as np
import matplotlib.pyplot as plt

# Given data points (replace with your dataset)
x = np.array([1, 2, 3, 4, 5])
y = np.array([2.7, 7.4, 20.1, 54.6, 148.4])  # Exponential trend

# Step 1: Apply Natural Logarithm
ln_y = np.log(y)

# Step 2: Compute necessary summations
n = len(x)
sum_x = np.sum(x)
sum_ln_y = np.sum(ln_y)
sum_x_ln_y = np.sum(x * ln_y)
sum_x2 = np.sum(x ** 2)

# Step 3: Calculate b
b = (n * sum_x_ln_y - sum_x * sum_ln_y) / (n * sum_x2 - sum_x ** 2)

# Step 4: Calculate a
A = (sum_ln_y - b * sum_x) / n
a = np.exp(A)

# Step 5: Generate fitted values
x_fit = np.linspace(min(x), max(x), 100)
y_fit = a * np.exp(b * x_fit)

# Step 6: Plot the original data and fitted curve
plt.scatter(x, y, color='red', label='Original Data')
plt.plot(x_fit, y_fit, color='blue', label='Exponential Fit')
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.title("Exponential Curve Fitting (Manual Calculation)")
plt.show()

# Print the computed values of a and b
print(f"Computed a: {a:.4f}, Computed b: {b:.4f}")
```
