```python
import matplotlib.pyplot as plt
import numpy as np
import math

n = int(input("Enter the number of data points: "))
x = []
y = []
for i in range(n):
    x.append(float(input(f"Enter x[{i+1}]: ")))
    y.append(float(input(f"Enter y[{i+1}]: ")))

x = np.array(x)
y = np.array(y)

ln_y = np.log(y)
sum_x = np.sum(x)
sum_ln_y = np.sum(ln_y)
sum_x_ln_y = np.sum(x * ln_y)
sum_x_squared = np.sum(x**2)

a1 = (n * sum_x_ln_y - sum_x * sum_ln_y) / (n * sum_x_squared - sum_x**2)
a0 = (sum_ln_y - a1 * sum_x) / n
a = np.exp(a0)

x_fit = np.linspace(min(x), max(x), 100)
y_fit = a * np.exp(a1 * x_fit)

plt.scatter(x, y, color='red', label='Original Data')
plt.plot(x_fit, y_fit, color='blue', label=f'Exponential Fit: y = {a:.2f} * e^({a1:.2f}x)')
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.title("Exponential Curve Fitting")
plt.show()

print(f"Equation of best fit: y = {a:.4f} * e^({a1:.4f}x)")
```
