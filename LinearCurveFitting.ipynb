```python
import numpy as np
import matplotlib.pyplot as plt

n = int(input("Enter the number of data points: "))
x = []
y = []
for i in range(n):
    x.append(float(input(f"Enter x[{i+1}]: ")))
    y.append(float(input(f"Enter y[{i+1}]: ")))

x = np.array(x)
y = np.array(y)

sum_x = np.sum(x)
sum_y = np.sum(y)
sum_xy = np.sum(x * y)
sum_x_squared = np.sum(x**2)

a = (n * sum_xy - sum_x * sum_y) / (n * sum_x_squared - sum_x**2)
b = (sum_y - a * sum_x) / n

y_pred = a * x + b

plt.scatter(x, y, color='blue', label="Actual Data")
plt.plot(x, y_pred, color='red', label=f"Fitted Line: y = {a:.2f}x + {b:.2f}")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Manual Linear Curve Fitting")
plt.legend()
plt.show()

print(f"Equation of best fit line: y = {a:.2f}x + {b:.2f}")
```
