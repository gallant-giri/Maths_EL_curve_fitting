```python
import numpy as np
import matplotlib.pyplot as plt

# Sample Data
x = np.array([1, 2, 3, 4, 5])
y = np.array([2.1, 2.9, 4.2, 5.1, 6.9])
```

```python
# Compute required summations
n = len(x)  # Number of data points
sum_x = np.sum(x)
sum_y = np.sum(y)
sum_xy = np.sum(x * y)
sum_x_squared = np.sum(x**2)

# Display results
print(f"n = {n}")
print(f"Σx = {sum_x}, Σy = {sum_y}, Σxy = {sum_xy}, Σx² = {sum_x_squared}")
```

```python
# Compute slope (a) and intercept (b)
a = (n * sum_xy - sum_x * sum_y) / (n * sum_x_squared - sum_x**2)
b = (sum_y - a * sum_x) / n

# Display results
print(f"Slope (a) = {a}")
print(f"Intercept (b) = {b}")
```

```python
# Compute predicted y values
y_pred = a * x + b

# Display results
print("Predicted y values:", y_pred)
```

```python
# Plot data points
plt.scatter(x, y, color='blue', label="Actual Data")

# Plot the fitted line
plt.plot(x, y_pred, color='red', label=f"Fitted Line: y = {a:.2f}x + {b:.2f}")

# Labels & Legend
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Manual Linear Curve Fitting")
plt.legend()
plt.show()
```
