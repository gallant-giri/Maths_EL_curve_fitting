```python
import numpy as np
import matplotlib.pyplot as plt

# Sample Data (You can replace it with your own dataset)
x = np.array([-2, -1, 0, 1, 2, 3])
y = np.array([4, 1, 0, 1, 4, 9])  # Parabolic shape


n = len(x)  # Number of data points
n
```

```python
# Compute Summations
sum_x = np.sum(x)
sum_x2 = np.sum(x**2)
sum_x3 = np.sum(x**3)
sum_x4 = np.sum(x**4)

sum_y = np.sum(y)
sum_xy = np.sum(x * y)
sum_x2y = np.sum(x**2 * y)

# Display calculated values
print(f"Σx = {sum_x}, Σx² = {sum_x2}, Σx³ = {sum_x3}, Σx⁴ = {sum_x4}")
print(f"Σy = {sum_y}, Σxy = {sum_xy}, Σx²y = {sum_x2y}")
```

```python
# Creating the system of equations matrix
A = np.array([[n, sum_x, sum_x2], 
              [sum_x, sum_x2, sum_x3], 
              [sum_x2, sum_x3, sum_x4]])

B = np.array([sum_y, sum_xy, sum_x2y])

# Solving for coefficients a0, a1, a2
a0, a1, a2 = np.linalg.solve(A, B)

# Displaying results
print(f"Computed coefficients: a0 = {a0}, a1 = {a1}, a2 = {a2}")
```

```python
# Generating predicted y values
x_fit = np.linspace(min(x), max(x), 100)  # Creating smooth curve
y_fit = a0 + a1*x_fit + a2*x_fit**2
y_fit
```

```python
x_fit
```

```python
plt.scatter(x, y, color='red', label='Original Data')
plt.plot(x_fit, y_fit, color='blue', label='Fitted Curve')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Polynomial Curve Fitting (Quadratic)')
plt.legend()
plt.show()
```
