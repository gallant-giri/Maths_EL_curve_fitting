## Least Squares method

Method of least squares is a parameter estimation method based on minimizing the sum of the squares of the residuals (a residual being the difference between an observed value and the fitted value provided by a model).


### Steps in methods of least squares

1. Data will be given, the objective is to find parameters of a hypothesis curve (assumed hypothesis function).
2. Assume a relationship between data 
3. Formulate normal equation 
4. Find parameters of the curve that best fits the data (which minimizes the squares of residuals, i.e., total error).

### Ordinary Least Squares Example

Linear regression model, uses a linear function as a hypothesis function.

#### Step 1: Assume a linear hypothesis function, $\hat{y}$

Here, we will assume that the best fit curve is a linear curve. i.e., $\hat{y} = mX + c$, where $m$ is the slope of the curve and $c$ is the y - intercept.
The objective is to find values of $m$ and $c$ that best fits the given data.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Sample data
X = np.array(
    [5, 12, 18, 25, 33, 40, 48, 55, 63, 70, 78, 85, 93, 100, 108, 115, 123, 130, 138, 145])
y = np.array(
    [3.5, 4.8, 7.2, 9.0, 11.5, 13.2, 12.8, 16.5, 18.1, 17.4, 21.0, 19.8, 25.2, 22.5, 27.4, 26.9, 31.5, 30.2, 35.8, 34.0])
```

#### Step 2: Derive the Normal Equation

Normal equations for linear function will be: 
$$ \sum{y} = m\sum{x} + cn $$ and 
$$ \sum{xy} = m\sum{x^2} + c\sum{x}$$

Solving for $m$ and $c$ will give, 

$$ m = \frac{\sum{xy} - n\bar{x}\bar{y}}{\sum{x^2} - n\,\bar{(x)^2}}$$ <br> 
$$c = \bar{y} - m\,\bar{x}$$

```python
n = len(X)
x_mean = np.mean(X)
y_mean = np.mean(y)
```

```python
numerator = np.sum(X * y) - n * x_mean * y_mean
denominator = np.sum(X * X) - n * x_mean * x_mean
a_opt = numerator / denominator
b_opt = y_mean - a_opt * x_mean

print(f"Fitted parameters: a = {a_opt:.4f}, b = {b_opt:.4f}")
```

```python
y_fit = a_opt * X + b_opt

# Plot the data and the fitted curve
plt.scatter(X, y, label="Data", color="red")
plt.plot(X, y_fit, label=f"Fitted line: y={a_opt:.2f}x+{b_opt:.2f}", color="blue")
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.title("Curve Fitting using Gradient Descent (Without SciPy)")
plt.show()
```
